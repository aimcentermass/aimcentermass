```python
"""
If a list of dict/series is passed and the keys are all contained in the DataFrames index, 
the order of the columns in the resulting DataFrame will be unchanged.
Iteratively appending rows to a DataFrame can be more computationally intensive than a single concatenate. 
A better solution is to append those rows to a list.
And then concatenate the list with the original DataFrame all at once.
Examples
"""
df = pd.DataFrame([[1, 2], [3, 4]], columns=list('AB'))
print('df')
print(df)
df2 = pd.DataFrame([[5, 6], [7, 8]], columns=list('AB'))
print('df2')
print(df.append(df2))
print('df now appended with ignore as true')
df.append(df2, ignore_index=True) # ignore set to true
```

```python
"""
The following, while not recommended methods for generating DataFrames, 
show two ways to generate a DataFrame from multiple data sources.
"""
# Less efficient:
df = pd.DataFrame(columns=['A'])
print('this is df')
for i in range(5):
    df = df.append({'A': i}, ignore_index=True)
print(df)

# More efficient:
df = pd.concat([pd.DataFrame([i], columns=['A']) for i in range(5)], ignore_index=True)
df
```

```python
merged = pd.merge(df1,df3, on='Year')
print(merged)
# missing years 2002 and 2005 but years are in the middle
```

```python
merged = pd.merge(df1,df3, on='Year')
merged.set_index('Year', inplace=True)
print(merged)
# now totally missing 2002 and 2005
```

```python
merged = pd.merge(df1,df3, on='Year', how='left') # df1 is left, so keeping its years
merged.set_index('Year', inplace=True)
print(merged)
# missing 2005
```

```python
merged = pd.merge(df1,df3, on='Year', how='right') # df3 is right, so keeping its years
merged.set_index('Year', inplace=True)
print(merged)
# missing 2002
```

```python
merged = pd.merge(df1,df3, on='Year', how='outer') # outer ismost inclusive
merged.set_index('Year', inplace=True)
print(merged)
# missing data in certain columns - see NaN - Not a number
```

```python
merged = pd.merge(df1,df3, on='Year', how='inner') # this is the default - for this the data must be joined
merged.set_index('Year', inplace=True)
print(merged)
# missing 2002, 2005
```

```python
"""https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.loc.html
"""
df = pd.DataFrame([[1, 2], [4, 5], [7, 8]],
                  index=['cobra', 'viper', 'sidewinder'],
                  columns=['max_speed', 'shield'])
print('the index is specifically called out for this dataframe:')
print(df, '\n')
print('Single label. Note this returns the row as a Series:')
print(df.loc['viper'], '\n')  # Single label. Note this returns the row as a Series
print('List of labels. Note using [[]] returns a DataFrame:')
print(df.loc[['viper', 'sidewinder']], '\n')  # List of labels. Note using [[]] returns a DataFrame
print(df.loc['cobra', 'shield'], '<- see it is the value at row cobra, clm shield.')
print(df.loc['cobra':'viper', 'max_speed'], 'max speed for cobra and viper.')
print(df.loc[[False, False, True]], 'done as a boolean.')
print(df.loc[df['shield'] > 6], 'conditional that returns a boolean series')
print(df.loc[df['shield'] > 6, ['max_speed']], 'conditional that returns a boolean with column labels')
print(df.loc[lambda df: df['shield'] == 8], 'Callable that returns a boolean Series')
df.loc[['viper', 'sidewinder'], ['shield']] = 50 # I don't think you can print this
print(df, 'Set value for all items matching the list of labels')
df.loc['cobra'] = 10
print(df,'Set value for an entire row, in this instance cobra to 10')
df.loc[:, 'max_speed'] = 30
print(df, 'set value for an entire column, in this instance max_speed to 30')
df.loc[df['shield'] > 35] = 0
print(df, 'Set all values for rows matching callable condition to a particular column value, here > 35 to 0')

```
